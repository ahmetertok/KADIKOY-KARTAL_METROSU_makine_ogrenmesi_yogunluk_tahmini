```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import folium as folium
```

```python
# read csv
df = pd.read_csv(
    'data/Ya_ Grubuna Gre Rayl_ Sistemler _stasyon Bazl_ Yolcu ve Yolculuk Say_lar_.csv')
# drop _id, passage_count, and _columns
df
```

```python
df = df.drop(columns=['_id', 'PASSAGE_COUNT',
             'STATION_NUMBER', 'TOWN', 'age_group', 'GENDER'])
df
```

```python
new_df = df[df['LINE'].str.contains('KADIKÖY-KARTAL METROSU')]
new_df
```

```python
new_df = new_df[new_df['transaction_Year'].astype(str).str.contains('2020')]
new_df
```

```python
new_df2 = new_df.groupby(by="STATION_NAME")["PASSENGER_COUNT"].sum()
# sort by passenger count
new_df2 = new_df2.sort_values(ascending=False)
new_df2
```

```python
# sort alphabetically by station name
new_df2 = new_df2.sort_index()
new_df2
```

```python
new_df2.to_csv('/Users/utku/ibb/metroline/newdf2.csv')
```

```python
# drop duplicates stations names
new_df = new_df.drop_duplicates(subset='STATION_NAME', keep='first')
new_df.info()
```

```python
#sort alphabetically by station name
new_df = new_df.sort_values(by='STATION_NAME')
```

```python
# add new_df2 to new_df as a column
new_df['PASSENGERS'] = new_df2.values
new_df
```

```python
# sort by passenger count
new_df = new_df.sort_values(by='PASSENGERS', ascending=False)
new_df
```

```python
# drop transactipon_month and passenger_count columns
new_df = new_df.drop(columns=['transaction_Month', 'PASSENGER_COUNT'])
```

```python
new_df.to_csv('/Users/utku/ibb/metroline/newdf.csv')
```

```python
latitude, longitude, station, passenger = new_df['LATITUDE'], new_df[
    'LONGITUDE'], new_df['STATION_NAME'], new_df['PASSENGERS']
locations = list(zip(latitude, longitude, station, passenger))
m = folium.Map(location=[40.9, 29],
               width="%100",
               height="%100",
               zoom_start=11)

for lat, lng, st, passe in locations:
    if passe > 1000000:
        folium.Circle(location=[lat, lng], radius=800,
                      popup=st, color='black', fill=True).add_to(m)
    if 1000000 > passe > 500000:
        folium.Circle(location=[lat, lng], radius=500,
                      popup=st, color='red', fill=True).add_to(m)
    if 500000 > passe > 100000:
        folium.Circle(location=[lat, lng], radius=300,
                      popup=st, color='orange', fill=True).add_to(m)
    if 5000 > passe > 1000:
        folium.Circle(location=[lat, lng], radius=200,
                      popup=st, color='yellow', fill=True).add_to(m)
m
```
